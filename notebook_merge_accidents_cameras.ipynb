# Merge Accidents and Nodes, then union with fixed camera locations\

### Import modules 

```python
import pandas
import datetime
```

### Set Paths 

```python
nodes_path = "/users/danielcorcoran/desktop/github_repos/tableau_projects/Vic Crash Stats (Extracted 4-5-2018)/raw data/NODES_DPC_2.csv"
accident_path = "/users/danielcorcoran/desktop/github_repos/tableau_projects/Vic Crash Stats (Extracted 4-5-2018)/raw data/ACCIDENT.csv"
camera_path = "/users/danielcorcoran/desktop/github_repos/tableau_projects/Vic Crash Stats (Extracted 4-5-2018)/raw data/Fixed Camera/fixed_camera_locations_with_lga.csv"
```

### Read in data

```python
nodes_data = pandas.read_csv(nodes_path)
```

```python
accident_data = pandas.read_csv(accident_path, low_memory=False)
```

```python
camera_data = pandas.read_csv(camera_path)
```

### Preview Data

```python
camera_data_subset = camera_data[["location_long", "location_short", "longitude", "latitude", "lga_code"]]

camera_data_subset.head()
```

```python
camera_data_subset.columns = ["camera_location_desc_long", "camera_location_desc_short", "Long", "Lat", "lga_code"]
```

### Merge Accident Data with Nodes data

```python
accident_nodes = pandas.merge(left=accident_data, right = nodes_data, how= "left", on="ACCIDENT_NO")
```

```python
accident_nodes.head()
```

```python
sorted(accident_nodes.columns)
```

### Add columns to each dataset to identify whether its camera or crash data

```python
accident_nodes["dataset_identifier"] = "Vic Crash Stats Data"
camera_data_subset["dataset_identifier"] = "Fixed Cameras Data"
```

```python
accident_nodes.head()
```

```python
camera_data_subset.head()
```

### Union vic crash data with fixed camera data

```python
final_data = pandas.concat([accident_nodes, camera_data_subset] , axis = 0, sort = True)
```
