## Add Short month columns and year column 

```python
for index in range(data.shape[0]):
    current_period = data.loc[index, "Period"].strip()
    csi = data.loc[index, "CSI"]
    if csi < 20:
        data.loc[index, "CSI bracket"] = "Totally dissatisfied"
    elif csi < 40:
        data.loc[index, "CSI bracket"] = "Very dissatisfied"
    elif csi < 60:
        data.loc[index, "CSI bracket"] = "Somewhat dissatisfied"
    elif csi < 80:
        data.loc[index, "CSI bracket"] = "Somewhat satisfied"
    elif csi < 100:
        data.loc[index, "CSI bracket"] = "Very satisfied"
    else:
        data.loc[index, "CSI bracket"] = "Totally satisfied"
        
        
    month_short = current_period[:3]
    year_short = current_period[-2:]
    year_full = "20" + year_short
    
    date_full = "13-" + month_short + "-" + year_full
    
    
    data.loc[index, "year_full"] = year_full
    data.loc[index, "month_short"] = month_short
    data.loc[index, "date_full"] = date_full
    
```
